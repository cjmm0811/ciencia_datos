```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
```

```python
url = "https://view.e.iryo.eu/?qs=24ca09354e1a3424f681a88dad4538ae2bd97c35fe5b248abe18df52997f6979039ce7328eea727c2e45b0e18ff54559df4beb5d0bbdc7bc4668ef3913385995efa40cf3567f56c6f487d2d4df1e17f4"
headers = {'User-Agent': 'Mozilla/5.0'}
```

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize
```

```python
req = Request(url, headers=headers)
raw_web = urlopen(req).read()

soup = BeautifulSoup(raw_web, "html.parser")
```

```python
soup
```

```python
spam = soup.find_all("tr")
len(spam)

spam
```

```python
text_to_tokenize = ' '.join([spams.text for spams in spam])
```

```python
tokens = word_tokenize(text_to_tokenize, "spanish")
```

```python
tokens = [word.lower() for word in tokens if word.isalpha()]
tokens
```

```python
from nltk.corpus import stopwords
```

```python
tokens_clean = tokens[:]

for token in tokens:
    if token in stopwords.words('spanish'):
        tokens_clean.remove(token)

tokens_clean
```

```python
from nltk.stem import SnowballStemmer
```

```python
spanish_stemmer = SnowballStemmer('spanish')
```

```python
stem_tokens = []

for token in tokens_clean:
    stem_tokens.append(spanish_stemmer.stem(token))

stem_tokens
```

```python
import spacy
nlp = spacy.load('es_core_news_sm')
```

```python
lem_tokens = []

for token in nlp(' '.join(tokens_clean)):
    lem_tokens.append(token.lemma_)

lem_tokens
```

```python
conteo = {}

for token in lem_tokens:
    if token in conteo.keys():
        conteo[token] += 1
    else:
        conteo[token] = 1

    if len(token) > 7:
        print(f"{token}, Esto es spam")
    else:
        print(f"{token}, Esto no es spam")


conteo
```
