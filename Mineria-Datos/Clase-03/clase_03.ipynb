```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
```

```python
url = "https://es.wikipedia.org/wiki/Anne_Hathaway"
headers = {'User-Agent': 'Mozilla/5.0'}
```

```python
req = Request(url, headers=headers)
raw_web = urlopen(req).read()
```

```python
soup = BeautifulSoup(raw_web, "html.parser")
```

```python
soup
```

```python
tables = soup.find_all("table", attrs={"class":"wikitable sortable"})
table = tables[0]

table
```

```python
table_headers = []
for col in table.find_all("th")[1:]: #Se pone el [1:] para quitar el primer valor ya que no lo necesitamos
    table_headers.append(col.text.strip())
```

```python
table_body_raw = table.find_all("tr")[2:]
```

```python
table_body = []

i = 0

for row in table_body_raw:

    template = []
    cols = row.find_all('td')
    

    if len(cols) == 3:
        template.append(table_body[i-1][0])

        for col in cols:
            template.append(col.text.strip())

    elif len(cols) == 2:
        template.append(table_body[i-1][0])
        template.append(cols[0].text.strip())
        template.append(table_body[i-1][2])
        template.append(cols[1].text.strip())

    elif len(cols) == 4:
        for col in cols:
            template.append(col.text.strip())

    if len(template) == 4:
        table_body.append(template)
        i += 1
```
