```python

import requests
import json

import pandas as pd
```

```python
url = "https://servicios.ine.es/wstempus/jsCache/es/DATOS_TABLA/33791?tip=AM"
```

```python
response = requests.get(url).text
objeto = json.loads(response)
```

```python

pre_df = []

for registro in objeto[:20]:
    template01 = {"id": registro['COD']}
    dict_data = {}
    dict_data.update(template01)

    for origen in registro['MetaData']:
        
        if origen['Variable']['Nombre'] not in ['Tipo de dato', 'Sexo']:  # Condici칩n antes de agregar al diccionario
            
            if origen['Variable']['Codigo']== '':
                template02 = {"Origen": origen['Nombre']}
                dict_data.update(template02)
                
            else:
                template03 = {"Poblacion": origen['Nombre']}
                dict_data.update(template03)
    for anios in registro['Data']:
        template04 = {anios['Anyo']: anios['Valor']}
        dict_data.update(template04)

    pre_df.append(dict_data)


```

```python
df = pd.DataFrame(pre_df)

df
```
